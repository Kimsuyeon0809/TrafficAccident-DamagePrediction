```r
train1 <- subset(train, select=-경상부상자여부)
head(train1)
```

```r
train_set <- sample_frac(train1, 0.7)
test_set <- setdiff(train1, train_set) #setdiff: 차집합

head(train_set)
```

로지스틱회귀 모델링 - 사망중상자여부

```r
m <- lm(formula = 사망중상자여부 ~ ., data = train_set)
summary(m)
```

```r
print("backward model mback")
mback <- step(m, direction = "backward")
```

예측

```r
head(test_set)
```

```r
predict_value <- predict(mback, test_set, type = "response") %>% tibble(predict_value= .)
predict_value %>% show()
```

```r
predict_df <- test_set %>% select(사망중상자여부) %>% dplyr::bind_cols(., predict_value)
predict_df %>% show()
```

```r
predict_df <- predict_df %>%
  mutate(predict_사망중상자여부 = as.factor(ifelse(predict_value > 0.5, 1, 0)))

predict_df %>% show()
```

ROC curve와 AUC

```r
install.packages("pROC")
```

```r
predict_df$사망중상자여부 <- as.factor(predict_df$사망중상자여부)
predict_df %>% show()
```

```r
roc_c <- pROC::roc(predict_df$사망중상자여부 , predict_df$predict_value )
pROC::plot.roc(roc_c,
              col = " royalblue",
              print.auc=TRUE,
              max.auc.polygon=TRUE,
              print.thres=TRUE, print.thres.pch=19, print.thres.col="red",
              auc.polygon=TRUE, auc.polygon.col="#D1F2EB")
```

```r
pROC::coords(roc_c, "best", ret="threshold", transpose=F)
```

```r
predict_cutoff_roc <- predict_df %>% mutate(predict_사망중상자여부 = as.factor(ifelse(predict_value > 0.2600063, 1, 0)))

predict_cutoff_roc %>% show()
```

```r
predict_value_mback <- predict(m, test_set, type = "response") %>%
 tibble(predict_value = .)

predict_check_mback <- test_set %>%
 select(사망중상자여부) %>%
 dplyr::bind_cols(., predict_value_mback)

roc_c_mback <- pROC::roc(predict_check_mback$사망중상자여부, predict_check_mback$predict_value)
```
